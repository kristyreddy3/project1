# drop unnamed column

```python
df = df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender",palette=['pink','skyblue'])
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution" ,color="purple")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
#sns.heatmap(gb)
sns.heatmap(gb,cmap=sns.cubehelix_palette(),annot=True)
plt.title("Relationship Between Parent's Eduaction and Student's Score", color="purple")
plt.show()
```

```python
#from the above chart we have cuncluded that the education of the parents have good impact on their students
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,cmap=sns.cubehelix_palette(),annot=True)
plt.title("students score due to their parents marital status",color="brown")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on
#the students score due to their parents marital status
```

# distribution of ethnic group

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels=l,autopct="%1.2f%%",colors=sns.color_palette('coolwarm'))
plt.title("Distribution of Ethnic Groups", color="teal")

plt.show()
```
